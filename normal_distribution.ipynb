```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def plot_normal_distribution(mean=0, std=1, num_samples=1000, bins=50):
    # generate random samples
    samples = np.random.normal(mean, std, num_samples)

    # axis config
    X_MIN = -10
    X_MAX = 10
    Y_MAX = 0.8
    
    # histogram
    plt.figure(figsize=(10, 6))
    count, bins_edge, _ = plt.hist(samples, bins=bins, density=True, alpha=0.6, color='purple', edgecolor='black', label='Histogram', range=(X_MIN, X_MAX))
    
    # ideal probability density function
    xmin, xmax = plt.xlim()
    x = np.linspace(X_MIN, X_MAX, 200)
    pdf = stats.norm.pdf(x, mean, std)
    plt.plot(x, pdf, 'r-', lw=2, label=f'Normal PDF (μ={mean}, σ={std})')
    
    # add mean and ±1σ, ±2σ lines
    plt.axvline(mean, color='green', linestyle='--', lw=2, label=f'Mean = {mean}')
    plt.axvline(mean + std, color='orange', linestyle=':', lw=2, label=f'+1σ = {mean + std:.2f}')
    plt.axvline(mean - std, color='orange', linestyle=':', lw=2, label=f'-1σ = {mean - std:.2f}')
    plt.axvline(mean + 2*std, color='purple', linestyle='--', lw=1)
    plt.axvline(mean - 2*std, color='purple', linestyle='--', lw=1)

    plt.xlim(X_MIN, X_MAX)
    plt.ylim(0, Y_MAX)
    
    plt.title(f'Normal Distribution Explorer\nSamples: {num_samples} | Mean: {mean} | Std: {std}', fontsize=14)
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    print(f"Sample statistics:")
    print(f"  Mean       : {samples.mean():.3f} (theoretical: {mean})")
    print(f"  Std dev    : {samples.std():.3f} (theoretical: {std})")
    print(f"  95% of data should fall between {mean - 1.96*std:.2f} and {mean + 1.96*std:.2f}")
```

```python
# standard normal
plot_normal_distribution(mean=0, std=1)

# different parameters
plot_normal_distribution(mean=2, std=2, num_samples=5000)

# narrow distribution
plot_normal_distribution(mean=0, std=0.6, num_samples=5000)
```
